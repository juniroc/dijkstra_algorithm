```python
s = {}
s[1] = 0
gr = {}
stack = []
explored = []

gr = {}
gr_node = {}
with open(path_1, 'r') as f:
    for line in f:
        items = line.split()
        
        # edge of head 정의
        for i in range(1,len(items)):
            split_c = items[i].split(',')
            if int(items[0]) not in gr_node:
                gr_node[int(items[0])] = [int(split_c[0])]
            else :
                gr_node[int(items[0])].append(int(split_c[0]))

        # eage length 정의
        for i in range(1, len(items)):
            split_c_2 = items[i].split(',')
            gr[(int(items[0]),int(split_c_2[0]))] = int(split_c_2[1])
```

```python
# dijkstra algorithm 만들기
def dijkstra(node):
    
    # node의 edge들 stack에 쌓기
    for i in gr_node[node]:
        stack.append((node,int(i)))
    
    
    while stack:
        edge = stack.pop(0)
        
        # 노드 거리 정의
        path_len = s[edge[0]] + gr[edge]

        # 거리정의가 안되어 있는 경우
        if edge[1] not in s:
            s[edge[1]] = s[edge[0]] + gr[edge]
            
            # 스택에 edge 쌓기
            for j in gr_node[edge[1]]:
                stack.append((edge[1],j))
            
        # 기존보다 짧은 경우
        if path_len < s[edge[1]]:
            s[edge[1]] = path_len
            
            # 스택에 edge 쌓기
            for j in gr_node[edge[1]]:
                stack.append((edge[1],j))

        
    return s
```
